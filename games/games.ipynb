### Работа с названием столбцов

```python
#Меняем регистр столбцов
df.columns = df.columns.str.lower()
```

### Преобразуем данные в нужные типы

```python
#Меняем тип данных в year_of_release на datetime
df = df.dropna(subset=['year_of_release'])
#df['year_of_release'] = df['year_of_release'][df['year_of_release'].notna()].astype('int')
df['year_of_release'] = df['year_of_release'].astype('int')
df['year_of_release'] = pd.to_datetime(df['year_of_release'], format='%Y') 
df["year_of_release"] = df["year_of_release"].dt.year
```

```python
#Меняем тип данных в user_score на float
#В столбце вылез некий tbd, это такое же обозначение пропущеного значения как и nan, так что он подлежит замене, на тот же nan
#Сначала заменяем пропущенные значения (nan и tbd) на те которые не встречаются в колонке, например 11
df["user_score"] = df["user_score"].str.replace('tbd','11').fillna('11')
df["user_score"] = pd.to_numeric(df["user_score"], downcast='float')
#А теперь меняем 11 на nan  np.nan
df.loc[df['user_score'] == 11, 'user_score'] = np.nan
```

```python
#Меняем тип данных в critic_score на int
#Сначала заменяем пропущенные значения на те которые не встречаются в колонке, например 101
df["critic_score"] = df["critic_score"].fillna(101)
df["critic_score"] = df["critic_score"].astype('int')
#А теперь меняем 11 на nan  np.nan
df.loc[df['critic_score'] == 101, 'critic_score'] = np.nan
```

#### **rating**

```python
df['rating'].unique()
```

Чесно сказать я и не понял что это такое. Хорошо, что не забанен в гугле. С 1994 года ESRB (Entertainment Software Rating Board) присвает играм возврастной рейтинг. Сразу возникает гипотеза, что пропуски стоят у тех чья дата релиза раньше 1994 года. Проверим.

```python
#Присвоем пропускам значение NAN.
df["rating"] = df["rating"].fillna("NOT_NAN")
df_test = df[df['rating'] == 'NOT_NAN'].groupby(['rating','year_of_release'])\
.agg(games=('name', 'count')).sort_values(by='games',ascending=False)
df.loc[df['rating'] == 'NOT_NAN', 'rating'] = np.nan
df_test.head(10)
```

Что ж это было познавательно, но гипотеза не верна. 

Основной причиной пропусков в столбце **rating** является, то что рейтинг взят от ESRB, а это региональная структура. Вполне возможно, что другие игры получили возврастной рейтинг от других организаций (Есть несколько основных возрастных рейтингов: американская ESRB, европейская PEGI, немецкая USK, японская CERO, российская RARS, а также еще несколько региональных организаций). Отсюда и берутся пропуски.

Так или иначе, поставим заглушку.

```python
df['rating'] = df['rating'].fillna('Other')
```

## Проверяем гипотезы

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Предположим, что пользователям так же нравится игры для приставки Xbox One, как нравится игры и для PC. 

Н0: Средний пользовательский рейтинги платформы Xbox One равен рейтингу PC.

Н1: Среднее пользовательский рейтинги платформы PC не равен рейтингу Xbox One.

```python
x = true_df[(true_df['platform'] =='XOne') & (true_df['user_score'] >= 0)]['user_score']
y = true_df[(true_df['platform'] =='PC') & (true_df['user_score'] >= 0)]['user_score']

results = st.ttest_ind(x, y, equal_var= False)

alpha = .01

print('p-значение:', results.pvalue )

if (results.pvalue < alpha):
    print('Среднее пользовательский рейтинги платформы PC не равен рейтингу Xbox One.')
else:
    print('Средний пользовательский рейтинги платформы Xbox One равен рейтингу PC')
print('Проверка:', x.mean(), y.mean())
```

Нулевая гипотеза принята - средний пользовательский рейтинг игр для Xbox One и PC равны. 

### Средние пользовательские рейтинги жанров Action и Sports разные

Наша гипотеза заключается в том, что средняя пользовательская оценка жанра Action равна средней пользовательской оценке жанра Sports. 

Н0: Средний пользовательский рейтинги жанра Action равен рейтингу Sports.

Н1: Среднее пользовательский рейтинги жанра Action не равен рейтингу Sports.

```python
x = true_df[(true_df['genre'] =='Action') & (true_df['user_score'] >= 0)]['user_score']
y = true_df[(true_df['genre'] =='Sports') & (true_df['user_score'] >= 0)]['user_score']

results = st.ttest_ind(x, y, equal_var= False)

alpha = .01

print('p-значение:', results.pvalue )

if (results.pvalue < alpha):
    print('Среднее пользовательский рейтинги жанра Action не равен рейтингу Sports.')
else:
    print('Средний пользовательский рейтинги жанра Action равен рейтингу Sports')
print('Проверка:', x.mean(), y.mean())
```

Нулевая гипотеза отвергнута - средние пользовательские рейтинги двух таких разных жанров, как и предполагалось, равными быть не могут.
