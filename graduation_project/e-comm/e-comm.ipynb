**Описание проекта**
- Для интернет-магазина товаров для дома «Пока все ещё тут», необходимо выявить профили покупателей, а также сделать полноценный анализ товарного ассортимента. Поможем магазину стать лучше, а клиентам — обустроить дом своей мечты.

**Цель проекта**
- Определить какие товары входят в основной и дополнительный ассортимент, чтобы грамотно предлагать покупателям дополнительные товары и оптимизировать закупки (нет смысла закупать много дополнительного товара, когда нет основного).

**Задачи**
- Исследовательский анализ данных
- Анализ товарного ассортимента.
- Сформулировать и проверить статистические гипотезы

**Датасет** 

Колонки в ecommerce_dataset.csv :
- **date** — дата заказа;
- **customer_id** — идентификатор покупателя;
- **order_id** — идентификатор заказа;
- **product** — наименование товара;
- **quantity** — количество товара в заказе;
- **price** — цена товара.

Датасет описывает транзакции интернет-магазина товаров для дома и быта «Пока все ещё тут».

## Предобработка 

### Работа со временем 

Приведем данные по времени в читаемый вид и добавим три дополнительных столбца 

```python
df['date'] = pd.to_datetime(df['date'], format = '%Y%m%d%H')
df['date_month'] = df['date'].astype('datetime64[M]')
df['date_week'] = df['date'].dt.day_name()
df['hours'] = df['date'].dt.hour
df['revenue'] = df.quantity * df.price
df.head(1)
```

### Дубликаты

Обзор данных не выявил полных дублей, но наверняка есть скрытые, проверим.

```python
#Приведем названия к общей форме 
df['product'] = df['product'].str.lower() #регистр
df['product'] = df['product'].str.replace('ё','е', regex=True) # заменим ё
df['product'] = df['product'].str.replace('(', '', regex=True) 
df['product'] = df['product'].str.replace(')', '', regex=True) 
df['product'] = df['product'].str.replace(',', '', regex=True) 
```

```python
x = df.duplicated().sum()  # проверка на отсутствие дубликатов
y = len(df[df[['product','customer_id', 'order_id']].duplicated() == True][['product', 'order_id']])

print('Количество полных дубликатов: {} строк'.format(x))
print('Количество частичных дубликатов: {} строк'.format(y))
print('Доля дубликатов от общего количества записей: {:.2%}'.format((x + y) / len(df))) 
```

```python
display(
    df[df[['product','customer_id', 'order_id']].duplicated() == True][['order_id', 'product', 'quantity']]
    .sort_values(by=['order_id', 'product'])
    .head(10)
)
```

Сисетма настроена так, что продукт в заказе в случае, если его больше 1 добавляет в колонку quantity (для чего она и создана). Тут же 28% датасета, к сожалению мы не можем узнать ошибка ли это при выгрузке или ошибка в логах, но эти данные предстоит отбросить, чтобы провести точное исследование. 

```python
df = df.drop_duplicates(['product', 'customer_id', 'order_id'])
```

### Работа с выбросами

```python
x = df['quantity'].quantile(.99) # перцентиль
y = df[df['quantity'] > x]['quantity'].count()

plt.figure(figsize=(20,5))
sns.boxplot(x='quantity', data=df) #работа с выбросами = boxplot
plt.title('Распределение количества товара в заказе по сырым данным.', fontsize=16)
plt.show()

# кол-во аномальных значений в столбце 'quantity'
print('99-й перцентиль кол-ва товара в заказе: {:.0f} '.format(x))
print('Количество строк с аномальными значениями: {} строк'.format(y))
print('Доля аномальных значений от общего количества записей: {:.1%}'.format(y / len(df))) 
```

Выбросы меньше процента, смело удаляем

```python
df = df[(df['quantity'] < x)]
```

Посмотрим теперь на покупателей и их количество заказов.

```python
# Подсчитаем количество заказов по покупателелям и посмотрим на результат.
x = df.groupby('customer_id', as_index=False).agg({'order_id':'nunique'})
x.columns = ['customer_id','orders']

y = x['orders'].quantile(.99) # перцентиль
z = x[x['orders'] > y]['orders'].count()

plt.figure(figsize=(20,5))
sns.boxplot(x='orders', data=x)
plt.title('Распределение количества заказов для каждого уникального покупателя.')
plt.show()

print('99-й перцентиль кол-ва заказов у покупателя: {:.0f} '.format(y))
print('Количество покупателей с аномальным числом заказов: {} '.format(z))
print('Доля аномальных покупателей от общего количества : {:.1%}'.format(z / len(df))) 
display(x[x['orders'] > y])
```

Посмтрим на самых подозрительных под индексом 707, 1046 и 1829

```python
z = x[x['orders'] > 16]['customer_id']
temp = df.query('customer_id in @z').groupby('customer_id').agg({'order_id':'nunique', 'quantity' : 'sum', 'revenue': 'sum'})
#Посмотрим насколько соответствует средний чек этих покупателей среднему по таблице
temp['mean_revenue'] = temp['revenue'] / temp['order_id']
temp['ok_revenue'] = (temp['mean_revenue'] / df.groupby('order_id')
                                                .agg({'quantity' : 'sum', 'revenue': 'sum'})['revenue']
                                                .mean()
                     )
display(temp)
print('Средняя цена одного заказа по датасету:', df.groupby('order_id').agg({'quantity' : 'sum', 'revenue': 'sum'})['revenue'].mean())
```

## Проверка гипотез

### Статистическая значимость различий в среднем чеке заказа между "Товары для сада и дачи" и "Товары для ванной комнаты, туалета и уборки"

- **Нулевая гипотеза**: Различия в среднем чеке заказа между категориями отсутствуют.
- **Альтернативная гипотеза**: Имеются различия в среднем чеке заказа между категориями.

```python
print("p-value = {0:.3f}".format(st.mannwhitneyu(df[df['category'] == 'Товары для сада и дачи']['revenue'], 
                                               df[df['category'] == 'Товары для ванной комнаты, туалета и уборки']['revenue'], 
                                               alternative = 'two-sided')[1]))
print('{0:.3f}'.format(df[df['category']=='Товары для сада и дачи']['revenue'].mean() /
                       df[df['category']== 'Товары для ванной комнаты, туалета и уборки']['revenue']
                       .mean() - 1))
```

p-value меньше 0,05 следовательно нулевая гипотеза не может быть отвергнута, относительное различие среднего чека между указанными группами составляет 64,2%.

### Статистическая значимость различий в среднем чеке заказа в 10:00 и в 14:00.

- **Нулевая гипотеза**: Различия в среднем чеке заказа в это время отсутствуют.
- **Альтернативная гипотеза**: Имеются различия в среднем чеке заказа в это время. 

```python
print("p-value = {0:.3f}".format(st.mannwhitneyu(df[df['hours'] == 10]['revenue'], 
                                               df[df['hours'] == 14]['revenue'], 
                                               alternative = 'two-sided')[1]))
print('{0:.3f}'.format(df[df['hours'] == 10]['revenue'].mean() /
                       df[df['hours'] == 14]['revenue']
                       .mean() - 1))
```

p-value больше 0,05 следовательно нулевая гипотеза отвергнута, относительное различие среднего чека между указанными группами составляет 12,9%.
